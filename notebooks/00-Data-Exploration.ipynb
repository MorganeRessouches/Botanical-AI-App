```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_DIR = '../data/split_ttv_dataset_type_of_plants/Test_Set_Folder' # https://www.kaggle.com/datasets/yudhaislamisulistya/plants-type-datasets
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
```

```python
test_dataset = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='int'
)
```

```python
class_names = test_dataset.class_names
print("Classes names:", class_names)
```

```python
images, labels = next(iter(test_dataset))
```

```python
random_index = np.random.randint(0, len(images))
```

```python
random_image = images[random_index]
random_label_index = labels[random_index]
class_name = class_names[random_label_index.numpy()]
```

```python
plt.figure(figsize=(8, 8))

plt.imshow(random_image.numpy().astype("uint8")) 
plt.title(f"Classe : {class_name}", fontsize=16)
plt.axis("off")
plt.show()
```
